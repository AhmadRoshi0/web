```python
%pip install nltk
import nltk 
from nltk.corpus import stopwords 
nltk.download( 'stopwords') 
nltk.download('punkt')
from nltk.tokenize import word_tokenize 
text = word_tokenize (listToStr) 
data = [word for word in text if not word in stopwords.words()]
print(data)
```

```python
data
```

```python
#import nltk
#nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer 
lema = WordNetLemmatizer() 
l =[]

for r in data:
       l.append(lema.lemmatize(r))
    

```

```python
print(lema.lemmatize('brown'))
```

```python
import re 
update= [ ] 
update=[re. sub('[^a-zA-Z0-9]+ ', '', _) for _ in l] 

```

```python
update
```

```python
# now we do clustring using sklearn library

from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
#import warnings filter 
from warnings import simplefilter

```

```python
#ignore all future warnings
simplefilter(action='ignore',category=FutureWarning)

v= TfidfVectorizer(stop_words= 'english')
y = v.fit_transform(update)
 
```

```python
cluster = 3
mdl = KMeans(n_clusters=cluster, init='k-means++', max_iter=100, n_init=1) 
mdl.fit(y)

print ("These are our clusters catagorized: ")
print (" ")

cent = mdl.cluster_centers_.argsort()[:,::-1] 
t = v.get_feature_names()
for i in range(cluster): 
        print (" ")
        print("Cluster %d:" % i)
        for i in cent[i, :10]:
                print( ' %s' % t[i])
                print (" ")
```

```python
print("Searching the word  :")
k = v.transform( ["identified"])
g= mdl.predict (k)
print(g)

k =v.transform ( ["claim that this cetral aspect of anarchism is definitive is to sell anarchism short"])
g = mdl.predict (k) 
print(g)
```
